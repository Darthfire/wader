```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina', quality=100)
```

```python
df = pd.read_csv('/content/train.csv')
df_Spanish = df.query('language == "Spanish"')
df_Spanish = df_Spanish.groupby(['label']).size()
df_Spanish.to_csv('Spanish-count.csv')
df2 = pd.read_csv('/content/Spanish-count.csv')
df2.columns = ['label', 'count']
plt.rcParams['figure.figsize'] = (10, 6)
fig, ax = plt.subplots()
x = df2['label']
ax.bar(
    x = df2['label'],
    height = df2['count'],
    width = 0.2,
    color = '#F1BF00'
    )
plt.xticks(np.arange(min(x), max(x)+0.5, 0.5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

ax.set_xlabel('Labels', labelpad=15, color='#333333', fontsize = 13)
ax.set_ylabel('Frequency Count', labelpad=15, color='#333333', fontsize = 13)
ax.set_title('Spanish Labels Frequency Count', pad=15, color='#333333',weight='bold', fontsize = 18)
plt.savefig("Spanish-Freq-Count.pdf", format="pdf", bbox_inches="tight")
fig.tight_layout()
```
